```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Generate synthetic dataset with correlated features
np.random.seed(0)
n = 1000

# Create three correlated features
x1 = np.random.normal(0, 1, n)
x2 = 2 * x1 + np.random.normal(0, 0.5, n)  # Correlated with x1
x3 = -1 * x1 + np.random.normal(0, 0.5, n)  # Negatively correlated with x1
x4 = np.random.normal(0, 1, n)  # Independent feature

# Create a DataFrame to hold the dataset
data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

# Step 2: Compute the correlation matrix
corr_matrix = data.corr()

# Step 3: Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
# Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate the dataset
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df = pd.DataFrame(data)

# Create 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(df['X'], df['Y'], df['Z'], color='blue', marker='o')

# Add title and labels
ax.set_title('3D Scatter Plot of Data Points')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

# Show the plot
plt.show()
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Display Bokeh plots in Jupyter notebook
output_notebook()

# Step 1: Generate x-values from 0 to 10
x = np.linspace(0, 10, 100)

# Step 2: Generate y-values as the sine of x
y = np.sin(x)

# Step 3: Create a Bokeh figure (using 'width' and 'height' instead of 'plot_width' and 'plot_height')
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)

# Step 4: Plot the sine wave
p.line(x, y, legend_label="sin(x)", line_width=2, color='blue')

# Step 5: Show the plot
show(p)
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
#markers based on the 'sizes' and 'colors' columns.



import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource

# Display Bokeh plots in the notebook
output_notebook()

# Step 1: Generate random data
np.random.seed(42)
n_points = 100
x = np.random.uniform(0, 100, n_points)
y = np.random.uniform(0, 100, n_points)
sizes = np.random.uniform(10, 50, n_points)  # Marker sizes
colors = np.random.randint(0, 256, n_points)  # Colors based on a numerical column

# Step 2: Create a ColumnDataSource to store the data
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Step 3: Create a color mapper
color_mapper = linear_cmap(field_name='colors', palette='Viridis256', low=min(colors), high=max(colors))

# Step 4: Create a Bokeh figure
p = figure(title="Bokeh Scatter Plot with Variable Sizes and Colors", x_axis_label='X', y_axis_label='Y', width=700, height=400)

# Step 5: Add a scatter plot to the figure
p.scatter('x', 'y', size='sizes', color=color_mapper, source=source, alpha=0.6, legend_label="Data Points")

# Step 6: Show the plot
show(p)
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset


from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6

# Display Bokeh plots in the notebook
output_notebook()

# Step 1: Create a dataset
fruits = ['Apples', 'Bananas', 'Pears', 'Oranges', 'Grapes', 'Peaches']
counts = [10, 20, 15, 25, 8, 12]

# Step 2: Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Step 3: Create a Bokeh figure for the bar chart
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label="Fruits", y_axis_label="Counts", height=400, width=700)

# Step 4: Add vertical bars to the plot
p.vbar(x='fruits', top='counts', width=0.5, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Step 5: Customize the plot
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Step 6: Show the plot
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Display Bokeh plots in the notebook
output_notebook()

# Step 1: Generate random data (e.g., normally distributed values)
data_hist = np.random.randn(1000)

# Step 2: Compute the histogram and bin edges
hist, edges = np.histogram(data_hist, bins=30)

# Step 3: Create a figure for the histogram
p = figure(title="Histogram of Randomly Generated Data", x_axis_label='Value', y_axis_label='Frequency', width=700, height=400)

# Step 4: Plot the histogram (using the edges and the histogram values)
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.6)

# Step 5: Show the plot
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
#data_heatmap = np.random.rand(10, 10) 
#x = np.linspace(0, 1, 10) 
#y = np.linspace(0, 1, 10) 
#xx, yy = np.meshgrid(x, y) 

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.palettes import Viridis256

# Display Bokeh plots in the notebook
output_notebook()

# Step 1: Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Step 2: Flatten the data for Bokeh plotting
x_flat = xx.flatten()
y_flat = yy.flatten()
data_flat = data_heatmap.flatten()

# Step 3: Create a ColumnDataSource for the data
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))

# Step 4: Create a color mapper for the heatmap
color_mapper = linear_cmap(field_name='value', palette=Viridis256, low=min(data_flat), high=max(data_flat))

# Step 5: Create a figure for the heatmap
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', tools='hover', tooltips=[("Value", "@value")],
           width=700, height=600)

# Step 6: Plot the heatmap using 'rect' glyph
p.rect(x='x', y='y', width=0.1, height=0.1, source=source, fill_color=color_mapper, line_color=None)

# Step 7: Add a color bar to the plot
color_bar = ColorBar(color_mapper=color_mapper['transform'], width=8, location=(0,0))
p.add_layout(color_bar, 'right')

# Step 8: Show the heatmap
show(p)
```
